# IMPORT LIBRARY

```python
import numpy as np
import math
```

# IMPORT DATA

```python
# X
input = np.array([
              [1,6,8,0],
              [33,2,3,5],
              [1,6,6,7],
              [2,9,8,4],
              [5,6,9,6],
              [32,13,0,2]
])
# y
output = np.array([
              [1,0,1,0],
              [1,0,1,0]
])
# W
weight = np.full((6,2),0.01).T
# b
bias = np.full((2,1),0.01)
```

```python
print("========================================")
print(input,input.shape)
print("========================================")
print(output,output.shape)
print("========================================")
print(weight,output.shape)
print("========================================")
print(bias,output.shape)
print("========================================")
```

Screen Shot 2565-03-28 at 22.35.48.png

```python
# logistic regression
# y = mx + c
y = np.dot(weight , input) + bias
y
```

# Function

```python
def sigmoid(Z):
  # use 1 record
  # z = 1 / (1 + math.exp(-y))
  # use multi record
  return 1 / (1 + np.exp(-Z))

```

```python
# sigmoid output more than 0.5 is 1
# sigmoid output less than 0.5 is 0
y_predict = sigmoid(y)
y_predict
```

```python
# https://github.com/codebasics/deep-learning-keras-tf-tutorial/blob/master/5_loss/loss_function_exercise_solution.ipynb
loss_mse = np.mean(np.square(y-y_predict))
loss_mse
```

```python
# talib
loss_mse = (1/6)*np.power(np.sum(y - y_predict),2)
loss_mse
```
